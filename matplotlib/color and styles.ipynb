```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lines = ax.plot(x,x**2, color='red',lw=2,marker='v')
lines[0].set_dashes([5,2,10,10,5,2])
ax.plot(x,x+2, color='black', ls = ':',label='X v X+2')
ax.legend()
```
