```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a**3
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig, axes = plt.subplots(ncols=2,nrows=1)

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig, axes = plt.subplots(ncols=2,nrows=2)

axes[0][1].plot(x,y)
axes[1][0].plot(a,b)
axes[1][1].set_title('BOT RIGHT')
# plt.tight_layout()
fig.suptitle('FIGURE LEVEL')
fig.subplots_adjust(wspace=.5,hspace=.5)
```
